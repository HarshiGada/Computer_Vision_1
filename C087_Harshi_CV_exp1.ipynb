```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data
```

```python
#image=data.camera()
#image=data.brick()
image=cv2.imread("/content/images.jpg")
```

```python
plt.imshow(image,cmap='gray')
```

```python
image.shape
```

```python
grad_x=cv2.Sobel(image,ddepth=cv2.CV_32F,dx=1,dy=0,ksize=3)
grad_y=cv2.Sobel(image,ddepth=cv2.CV_32F,dx=0,dy=1,ksize=3)
```

```python
plt.imshow(grad_x,cmap='gray')
```

```python
grad_x_abs=cv2.convertScaleAbs(grad_x)
```

```python
plt.imshow(grad_x_abs,cmap='gray')
```

```python

grad_y_abs=cv2.convertScaleAbs(grad_y)
plt.imshow(grad_y_abs,cmap='gray')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(image,cmap='gray')
plt.title('Original')
plt.subplot(1,3,2)
plt.imshow(grad_x_abs,cmap='gray')
plt.title(' Absolute Gradient X')
plt.subplot(1,3,3)
plt.imshow(grad_y_abs,cmap='gray')
plt.title('Absolute Gradient Y')
```

```python
edge_image=np.sqrt((grad_x**2)+(grad_y**2))
edge_image=edge_image/np.max(edge_image)
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(edge_image,cmap='gray')
plt.title('Edges')
```

```python
angle=np.arctan2(grad_y,grad_x)*180/np.pi%180
row, col = image.shape[:2]
```

```python
angleID=np.reshape(angle,(-1,1))
plt.hist(angleID,bins=180)
plt.xlabel('Orientation of angle')
plt.ylabel('Number of pixels')
plt.title('Orientation of edges')
```
